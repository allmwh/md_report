```python
import pandas as pd
```

```python
pka = pd.read_csv('1AO6_amber.pkout',delim_whitespace=True)  
```

```python
pka['Subtraction'] =0.00
```

```python
for i in pka.index:
    if pka["pK_(1/2)"][i]=='>12.000':
        a=float(pka["pKint"][i])-float(12)
    elif pka["pK_(1/2)"][i]=='<0.000':
        a=float(pka["pKint"][i])-float(0)
    else:
        a=float(pka["pKint"][i])-float(pka["pK_(1/2)"][i])
    pka["Subtraction"][i]=abs(a)
pka=pka.sort_values("Subtraction",ascending=False)
```
